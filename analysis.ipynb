# OBJECTIVE 3: Feature Engineering
Extract useful time-related features (Year, Month, Day, Weekday) from the Date column. Create new relevant business metrics, including:
- Discounted price
- Sell-through rate

```python
# Extract year and month from the Date column
df['Year']=df['Date'].dt.year
df['Month']=df['Date'].dt.month
```

```python
# Extract day and weekday name from the Date column
df['Day']=df['Date'].dt.day
df['Weekday']=df['Date'].dt.day_name()
```

```python
df
```

```python
# Calculate discounted price based on original price and discount percentage
df["Discounted Price"] = df["Price"] * (1-df["Discount"]/100)
```

```python
df
```

```python
# Calculate Sell Through Rate
df["Sell Through Rate"] = df["Units Sold"]/df["Inventory Level"]
```

```python
df.head(2)
```

```python
# Analyse distribution of Sell Through Rate
df["Sell Through Rate"].describe()
```

# OBJECTIVE 4: Aggregations & Business Insights
Analyse demand patterns across key business factors such as:
- Product category
- Region and Seasonality
- Promotion activity
Investigate monthly demand patterns and seasonal trends to understand how demand changes over time and across product categories.

```python
# Analyse demand by product category
df.groupby("Category")["Demand"].agg(["mean", "sum","std"]).sort_values(by="sum", ascending= False).reset_index()
```

```python
# Analyse average demand by region and seasonality
df.groupby(["Region", "Seasonality"])["Demand"].mean()
```

```python
# Compare average demand when promotions are active vs inactive
df.groupby("Promotion")["Demand"].mean()
```

```python
# Create a pivot table for demand across months and categories
pd.pivot_table(df, values = "Demand", index = "Month", columns = "Category", aggfunc = "mean")
```

# OBJECTIVE 5: Data Visualisation
Use visualisations to uncover patterns and potential predictors of demand in the dataset, including:
- Demand distribution
- Inventory vs. sales relationship
- Category demand variation
- Weather impact on demand

```python
# Plot histogram to understand distribution of demand values
sns.histplot(df["Demand"], bins=20, kde=True)
plt.title("Demand Distribution")
plt.show()
```

```python
# Scatter plot to examine relationship between inventory level and units sold
sns.scatterplot(data=df, x="Inventory Level", y="Units Sold")
plt.title("Inventory Level Versus Units Sold")
plt.show()
```

```python
# Compare demand distribution across product categories
sns.boxplot(data = df, x = "Category", y="Demand")
plt.xticks(rotation=45)
plt.title("Demand By Category")
plt.show()  
```

```python
# Compare demand across weather conditions
sns.boxplot(data = df, x = "Weather Condition", y="Demand")
plt.xticks(rotation=45)
plt.title("Demand By Weather Condition")
plt.show()  
```

```python
# Calculate average monthly demand
monthly_demand = df.groupby("Month")["Demand"].mean()
```

```python
monthly_demand
```

```python
# Plot average demand by month using bar chart
monthly_demand.plot(kind = "bar")
plt.title("Average Demand by Monts") 
```

```python
# Calculate daily demand trends
daily_demand = df.groupby("Date")["Demand"].mean()
```

```python
daily_demand
```

```python
# Plot demand trend over time
daily_demand.plot()
plt.title("Daily Demand Over Time")
plt.xlabel("Date")
plt.ylabel("Demand")
plt.show()
```

```python
# Plot a chart showing impact of promotions on demand
sns.barplot(data=df, x="Promotion", y ="Demand")
plt.title("Promotion Impact On Demand")
plt.show()
```

```python
# Visualise relationship between discounted price and demand
sns.regplot(data = df, x = "Discounted Price", y = "Demand")
plt.title("Discounted Price vs Demand")
plt.show()
```

```python
# Plot a chart showing average demand by seasonality
df.groupby("Seasonality")["Demand"].mean().plot(kind="bar", title="Demand By Seasonality")
plt.show()
```

```python
# Plot a chart showing average demand when epidemics are active vs inactive
df.groupby("Epidemic")["Demand"].mean().plot(kind="bar", title="Impact of Epidemic On Demand")
plt.show()
```
